```python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()
load_dotenv(override=True)
```

```python
openai_key=os.getenv("OPENAI_API_KEY")   
client=OpenAI(api_key=openai_key)
```

```python
# System prompt — defined once, before the loop
messages = [
    {"role": "system", "content": """You are a personal journal assistant. 
Follow these rules strictly:
- Respond in 2-3 sentences maximum
- When asked to pick one option, state the choice and one reason only
- Never use bullet points or numbered lists
- Never end with phrases like 'feel free to ask' or 'have a great day'
- If you must list things, write them as natural prose"""},
    {"role": "assistant", "content": "Hi! What's on your mind today?"}
]
turn_counter=0
def summarize(messages):
    summary_messages= messages.copy()
    summary_messages.append({
        "role": "user",
        "content": "Please summarize our conversation so far in a few bullet points."
    })
    response=client.chat.completions.create(
            model='gpt-4o-mini',
            messages=summary_messages

        )
    return response.choices[0].message.content
while True:
    user_input = input("You: ")
     # ← user role here

    
    if user_input.lower()=='quit':
        print('goodbye')
        break
    if user_input.lower()=='summary':
        print(summarize(messages))
        continue


    
    messages.append({"role":"user", "content": user_input}) 
    
    response = client.chat.completions.create(
        model="gpt-4o-mini", 
        messages=messages
        )
    reply=response.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})
    
    print(reply)
    turn_counter += 1
    if turn_counter%3==0:
        if turn_counter%3==0:
            print("\n--- Auto Summary ---")
            print(summarize(messages))
            print("--------------------\n")
        
        


```
